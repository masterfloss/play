```python
import requests
import pandas as pd
import playscript as ps

url = "https://raw.githubusercontent.com/masterfloss/play/refs/heads/main/data/play1.txt"

response = requests.get(url)

if response.status_code == 200:
    play_text = response.text
```

```python

analysis = ps.PlayAnalysis(play_text )
analysis.plot_subjectivity_evolution()
analysis.plot_polarity_evolution()
analysis.df=analysis.df
```

```python
network = ps.InteractionNetwork(analysis.df)
network.visualize_network()
```
